```python
from datetime import datetime
import pandas as pd
```

```python
# Задание 1
# Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
# Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.
```

```python
%%time
# version 1
df = (pd.read_csv('ml-latest-small/ratings.csv')
      .groupby('userId')
      .filter(lambda x: x['userId'].count() > 100)
      .groupby('userId')
      .agg({'userId': 'count', 'timestamp': ['min', 'max']})
      .reset_index(drop=True)
      .sort_values([('userId', 'count')]))
df['difference'] = df[('timestamp', 'max')] - df[('timestamp', 'min')]
datetime.utcfromtimestamp(df['difference'].mean()/1000).strftime('%H:%M:%S')
```

```python
%%time
# version 2
(datetime.utcfromtimestamp(pd.read_csv('ml-latest-small/ratings.csv')
                           .groupby('userId')
                           .filter(lambda x: x['userId'].count() > 100)
                           .groupby('userId')
                           .agg({'userId': 'count', 'timestamp': ['min', 'max']})
                           .reset_index(drop=True).sort_values([('userId', 'count')])
                           .apply(lambda row: row[('timestamp', 'max')] - row[('timestamp', 'min')], axis=1)
                           .mean()/1000).strftime('%H:%M:%S'))
```

```python
# Задание 2
# Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия).
# Необходимо сформировать две таблицы:
# таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
# аналогичную таблицу по типам выручки с указанием адреса клиента
# Обратите внимание, что в процессе объединения таблиц данные не должны теряться.
    # rzd - железнодорожные перевозки
    # auto - автомобильные перевозки
    # air - воздушные перевозки
    # client_base - адреса клиентов
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
joined_df = [df.set_index('client_id') for df in [client_base, auto, air, rzd]]
joined_df = joined_df[0].join(joined_df[1:])
joined_df
```

```python
# Таблица 1
joined_df.reset_index().melt(id_vars=['client_id'], value_vars= ['auto_revenue', 'air_revenue', 'rzd_revenue'], var_name='type_revenue').dropna().sort_values('client_id')
```
